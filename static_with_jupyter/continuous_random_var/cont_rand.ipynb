```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy import integrate #scipy 모듈에서 적분 import
import warnings

warnings.filterwarnings('ignore', category=integrate.IntegrationWarning)
```

# 연속형 확률 변수
- 확률변수가 취할 수 있는 값이 연속적인 확률 변수
- 특정 값을 취하는 확률은 정의되지 않음
- 확률변수가 어느 구간에 들어가는 확률을 정의

- 룰렛
    - 취할 수 있는 값이 0 부터 1사이의 실수
    - 큰 수일수록 나오기 쉬워지는 불공정한 구조
    - 0.5라는 값을 취할 확률은 0
    
### 확률밀도 함수
- 확률변수가 취할수 있는 값은 구간 [a, b]
- 확률은 확률밀도함수(PDF)또는 밀도함수 f(x)에 의해 정의
- 어떤 특정 값을 취하는 확률로는 정의되지 않음


```python
x_range = np.array([0, 1])
```

```python
def f(x):
    if x_range[0] <= x <= x_range[1]:
        return 2*x #0~1에서는 2x로 확률밀도함수를 정의한거야
    else:
        return 0
    
```

```python
X = [x_range, f]
```

```python
xs = np.linspace(x_range[0], x_range[1], 100)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.plot(xs, [f(x) for x in xs], label='f(x)', color='gray')
ax.hlines(0, -0.2, 1.2, alpha=0.3)
ax.vlines(0, -0.2, 2.2, alpha=0.3)
ax.vlines(xs.max(), 9, 2.2, linestyles= ':', color='gray')

#0.4부터 0.6까지의 x좌표를 준비
xs = np.linspace(0.4, 0.6, 100)

#xs의 범위로 f(x)와 x축으로 둘러싸인 영역에 색을 적용
ax.fill_between(xs, [f(x) for x in xs], label='prob')

ax.set_xticks(np.arange(-0.2, 1.3, 0.1))
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.2, 2.1)
ax.legend()

plt.show()
```

```python
#불공정한 룰렛에 대한 밀도함수와 확률
```

```python
integrate.quad(f, 0.4, 0.6)
```

```python
from scipy.optimize import minimize_scalar

res = minimize_scalar(f)

res.fun
```

```python
integrate.quad(f, -np.inf, np.inf)[0]
```

```python
# 누적 분포 함수
```

```python
def F(x):
    return integrate.quad(f, -np.inf, x)[0]
```

```python
F(0.6) - F(0.4)
```

```python
xs = np.linspace(x_range[0], x_range[1], 100)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.plot(xs, [F(x) for x in xs], label='F(x)', color='gray')
ax.hlines(0, -0.1, 1.1, alpha=0.3)
ax.vlines(0, -0.1, 1.1, alpha=0.3)
ax.vlines(xs.max(), 0, 1.1, linestyles= ':', color='gray')

ax.set_xticks(np.arange(-0.1, 1.2, 0.1))
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)
ax.legend()

plt.show()
```
